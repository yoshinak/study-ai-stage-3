## 1. 要点まとめ


### 用語

入力層から中間層への数式に使用する用語は以下の通り。

- 入力: $ x_i $
- 重み: $ w_i $
- バイアス: $ b $
- 総入力: $ u $
- 出力: $ z $
- 活性化関数: $ f $
- 入力層のインデックス: $ i $


### 数式

入力層から中間層に変換する数式は、以下の通り。
(入力層のインデックス: $ i = 4 $)

$$
W = \begin{bmatrix}
w_1 \\
\vdots \\
w_4
\end{bmatrix},
x = \begin{bmatrix}
x_1 \\
\vdots \\
x_4
\end{bmatrix}
$$

$$
u = w_1 x_1 + w_2 x_2 + w_3 x_3 + w_4 x_4 + b \\
= W x + b
$$


さらに、 $ u $ に活性関数を適用した以下が出力 $ z $ となり、次の層の入力になる。

$$
z = f(u)
$$
